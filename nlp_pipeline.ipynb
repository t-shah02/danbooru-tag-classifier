```python
import pandas as pd
import spacy
from tqdm import tqdm

nlp = spacy.load('en_core_web_md', disable=['tok2vec', 'ner'])
```

```python
tags_df = pd.read_csv('./data/danbooru_tags.csv')
tags_df.head(n=10)  
```

```python
lemmatize_docs = lambda docs: " ".join(doc.lemma_ for doc in docs)


lemmatized_tags = []
for docs in tqdm(nlp.pipe(tags_df['tags'], n_process=4), total=tags_df.shape[0]):
    lemmatized_tags.append(lemmatize_docs(docs))

tags_df['lemmatized_tags'] = lemmatized_tags

tags_df.head(n=10)
```

```python
final_tags_df = tags_df.drop(columns=['tags'], axis=1)
final_tags_df.to_csv("./data/final-tag-data.csv", index=False)
```
